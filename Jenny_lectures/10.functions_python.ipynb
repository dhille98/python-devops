### üîß What is a Function in Python?

A **function** in Python is a **reusable block of code** that performs a specific task. You define it once, and then you can call (use) it as many times as needed.

---

### ‚úÖ Why Use Functions?

- To avoid repeating code (DRY principle ‚Äì Don‚Äôt Repeat Yourself)
- To make your code modular and organized
- To improve readability and maintainability

---

### üß± **Function Syntax:**

```python
def function_name(parameters):
    # code block
    return result
```

---

### üìå **Example ‚Äì A Simple Function:**

```python
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")
greet("Bob")
```

üñ®Ô∏è Output:
```
Hello, Alice!
Hello, Bob!
```

### üîÅ **Function with Return Value:**

```python
def add(a, b):
    return a + b

result = add(5, 3)
print("Sum is:", result)
```

üñ®Ô∏è Output:
```
Sum is: 8
```
